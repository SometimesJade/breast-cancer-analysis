## Data analyse BDAAV
In dit Notebook wordt een dataset onderzocht met verschillende kenmerken van vrouwen die een borstkanker diagnose hebben gekregen. Er word onder andere gekeken naar de relatie tussen leeftijd, de grootte van de tumor en waar in de borst de tumor is gevonden. De dataset is [hier](https://www.kaggle.com/datasets/fatemehmehrparvar/breast-cancer-prediction) gevonden.

```python
# Importeren van alle benodigde libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import chi2_contingency
from scipy.stats import mannwhitneyu


# Het inlezen van de ruwe onbewerkte data in een DataFrame.
pd.set_option('future.no_silent_downcasting', True)
raw_data = pd.read_csv("breast-cancer-dataset.csv", index_col=0)

# Een voorbeeld van de eerste 5 regels uit de onbewerkte data.
raw_data.head(5)
```

### Beschrijving kolommen
De dataset is als volgens opgedeeld: 
<ol>
    <li>S/N: Identificatie index uit de dataset</li>
    <li>Age: Leeftijd in jaren</li>
    <li>Menopause: Of de patient voor of na de menopauze is</li>
    <li>Tumor Size (cm): De grootte van de tumor in centimeter</li>
    <li>Inv-Nodes: Of de kankercellen gevonden zijn in het lymfestelsel</li>
    <li>Breast: Welke borst de tumor is gevonden, links of rechts</li>
    <li>Metastasis: Of er metastasen heeft plaatsgevonden</li>
    <li>Breast Quadrant: In welk kwadrant van de borst de tumor gevonden is</li>
    <li>History: Of de patient/een familielid eerder een tumor heeft gehad</li>
    <li>Diagnosis Result: Het resultaat van de diagnose, of de tumor goedaardige (benign) of kwaadaardige (malignant) is</li>
</ol>
Alle data in de dataset is geformarteerd als string in de ruwe data, daarom is de data eerst opgeschoont voor deze gebruikt is

## De data opschonen
Als eerst word de data doorzocht naar missende waardes, de missende waardes zijn in de dataset aangegeven met een '#'.
De functie find_missing_value() bekijkt of kolom de waarde # bevat en toont alle gevonden waardes.
De index van de missende waarde wordt opgehaald en deze wordt toegevoegd aan een lijst met alle missende waardes.

```python
'''
    Deze functie vind de index van de foutieve waardes in het opgegeven dataframe
    :param df: A pandas dataframe
    :return: list
'''
def find_missing_values(df):
    missing_index = []
    for item in df:
        data_missing = df.index[df[item] == '#'].values
        if len(data_missing) > 0:
            print(item)
            print(f"Values found in column: {df[item].unique()}")
            print(f"Missing {item} value found at index: {data_missing}\n")
        
        for index in data_missing:
            if index not in missing_index:
                missing_index.append(int(index))
                
    if missing_index:
        print(f"Missing indexes found at: {missing_index}")
        return sorted(missing_index)
    else:
        print("No missing indexes found!")
        return None
    
# Het uitvoeren van de functie geeft de volgende output. 
missing_data = find_missing_values(raw_data)
```

## Working DataFrame aanmaken
Er is een kopie gemaakt van het DataFrame, en zijn de indexen met de missende waardes verwijderd. De functie find_missing_values() geeft nu geen gevonden missende waardes meer terug.

```python
breast_cancer_df = raw_data.copy()
breast_cancer_df = breast_cancer_df.drop(missing_data)

find_missing_values(breast_cancer_df)
```

## DataFrame formateren
In de dataset is nog één abnormale waarde te vinden. In de kolom "Inv-Nodes", is er een onderscheid gemaakt of de kankercellen ook in het lymfestelsel zijn gevonden. Deze onderscheiding in de dataset is gemaakt tusesn 0, niet verspreidt naar het lymfestelsel en 1, wel verspreidt naar het lymfestelsel. Er is in deze kolom één keer de waarde 3 gevonden. Deze is ook uit de dataset verwijdert. Vervolgens worden er twee nieuwe subset DataFrames aangemaakt, één met alle goedaardige diagnoses en één met alle kwaadaardige diagnoses. Dit maakt het makkelijker om op het resultaat van de diagnose te groeperen en filteren. Ook worden de nodige kolommen omgezet naar integers.

```python
group_by_nodes = breast_cancer_df.groupby("Inv-Nodes")
error_value = breast_cancer_df.index[breast_cancer_df["Inv-Nodes"] == '3'].values # Rij waar waarde 3 in gevonden is
breast_cancer_df = breast_cancer_df.drop(error_value) # Gevonden waarde uit df verwijderen


# Omzetten van waardes naar integers
breast_cancer_df.loc[:,"Tumor Size (cm)"] = pd.to_numeric(breast_cancer_df["Tumor Size (cm)"])
breast_cancer_df.loc[:,"Metastasis"] = pd.to_numeric(breast_cancer_df["Metastasis"])
breast_cancer_df.loc[:,"History"] = pd.to_numeric(breast_cancer_df["History"])
breast_cancer_df.loc[:,"Age"] = pd.to_numeric(breast_cancer_df["Age"])


# Het aanmaken van de bovengenoemde twee DataFrames 
benign_total = breast_cancer_df.query("`Diagnosis Result` == 'Benign'")
malignant_total = breast_cancer_df.query("`Diagnosis Result` == 'Malignant'")
```

## Leeftijd
Als eerst is er gekeken naar de overeenkmosten tussen de leeftijd en hoe vaak een goedaardige of kwaadaardige tumor gevonden is. Dit is weergegeven in een histogram en een dichtheidsplot. Zoals te zien is in de grafieken komen goedaardige tumoren vaker voor op een lagere leeftijd en kwaadaardige tumoren vaker op een hogere leeftijd.

```python
# Het aanmaken van het figuur met titel
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle("Age by Diagnosis", fontsize=14)

# De eerste grafiek (links) in het figuur
ax[0].hist(benign_total["Age"], bins=15, alpha=0.5, color="blue", label = "Benign")
ax[0].hist(malignant_total["Age"], bins=15, alpha=0.5, color="red", label = "Malignant")
ax[0].set_xlabel("Age in years")
ax[0].set_ylabel("Occurences")

# De tweede grafiek (rechts) in het figuur
sns.kdeplot(benign_total["Age"], color="blue", linewidth=2, fill=True, ax=ax[1])
sns.kdeplot(malignant_total["Age"], color="red", linewidth=2, fill=True, ax=ax[1])
ax[1].set_xlabel("Age in years")

# Het aanmaken en plaatsen van de legende
handles, labels = ax[0].get_legend_handles_labels()
fig.legend(handles, labels, loc="lower center", ncol=2)

# Het figuur weergeven
plt.tight_layout()
plt.show()

# U-test
# H0: Er is geen verschil in leeftijd tussen goed- en kwaardaardige tumoren
# H1: Er is een verschil in leeftijd tussen goed- en kwaardaardige tumoren
benign_age = benign_total["Age"].astype(int)
malignant_age = malignant_total["Age"].astype(int)

u_stat, p_value = mannwhitneyu(benign_age, malignant_age, alternative="two-sided")
print("p-value:", p_value)
print("u-stat:", u_stat)

# Effect size berekenen
n1, n2 = len(benign_age), len(malignant_age)
rank_biserial = 1 - (2 * u_stat) / (n1 * n2)
print("Rank-biserial correlation:", rank_biserial)
```

De Mann–Whitney U-test toonde een significant verschil in leeftijd tussen goedaardige en kwaadaardige tumoren (U = 1849, p = <0.0001). De rank-biserial correlatie (r_rb = 0.638) wijst op een groot effect, waarbij patiënten met kwaadaardige tumoren gemiddeld ouder waren.

## Tumor grootte
Er is gekeken naar de overeenkomsten tussen de grootte van de gevonden tumor en of de tumor goedaardige of kwaardaardig is. Zoals in de grafieken te zien is, is de tumor vaak kleiner wanneer de tumor goedaardig is.


```python
# Het aanmaken van het figuur met titel
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle("Tumor Size by Diagnosis", fontsize=14)

# De juiste series worden uit de DataFrames opgehaald, het tellen van de tumor groottes per DataFrame
group_by_size_benign = benign_total.groupby("Tumor Size (cm)")
benign_counts_size = group_by_size_benign["Tumor Size (cm)"].agg("count")
group_by_size_malignant = malignant_total.groupby("Tumor Size (cm)")
malignant_counts_size = group_by_size_malignant["Tumor Size (cm)"].agg("count")

# De eerste grafiek (links) in het figuur
ax[0].bar(benign_counts_size.index, benign_counts_size.values, alpha=0.5, color="blue", label = "Benign")
ax[0].bar(malignant_counts_size.index, malignant_counts_size.values, alpha=0.5, color="red", label = "Malignant")
ax[0].set_ylabel("Occurences")
ax[0].set_xlabel("Tumor Size (cm)")

# De tweede grafiek (rechts) in het figuur
sns.kdeplot(benign_total["Tumor Size (cm)"], color="blue", ax=ax[1], fill=True, linewidth=2)
sns.kdeplot(malignant_total["Tumor Size (cm)"], color="red", ax=ax[1], fill=True, linewidth=2)

# Het aanmaken en plaatsen van de legende
handles, labels = ax[0].get_legend_handles_labels()
fig.legend(handles, labels, loc="lower center", ncol=2)

# Het figuur weergeven
plt.tight_layout()
plt.show()

# U-test
# H0: Er is geen verschil in tumor grootte en tussen goed- en kwaardaardige tumoren
# H1: Er is een verschil in tumor grootte en tussen goed- en kwaardaardige tumoren
benign_size = benign_total["Tumor Size (cm)"].astype(int)
malignant_size = malignant_total["Tumor Size (cm)"].astype(int)

u_stat, p_value = mannwhitneyu(benign_size, malignant_size, alternative="two-sided")
print("p-value:", p_value)
print("u-statL", u_stat)

# Effect size berekenen
n1, n2 = len(benign_size), len(malignant_size)
rank_biserial = 1 - (2 * u_stat) / (n1 * n2)
print("Rank-biserial correlation:", rank_biserial)
```

De tumorgrootte verschilde significant tussen goedaardige en kwaadaardige tumoren (U = 947.5, p = <0.0001). De rank-biserial correlatie (r_rb = 0.814) wijst op een groot effect.

## Metastasen
Als volgende is er gekeken naar de verschillen tussen metastaseren van de kankercellen bij goedaardige en kwaardaardige tumororen. Zoals in grafieken te zien is, is metastase van een goedaardige tumor erg zeldzaam in deze dataset.

```python
# Het aanmaken van het figuur met titel
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle("Metastasis by Diagnosis", fontsize=14)

# Het tellen van het aantal metastasen per DataFrame
group_by_meta_benign = benign_total.groupby("Metastasis")
benign_counts = group_by_meta_benign["Metastasis"].agg("count")
group_by_meta_malignant = malignant_total.groupby("Metastasis")
malignant_counts = group_by_meta_malignant["Metastasis"].agg("count")

# Het aanmaken van een nieuwe x-as
x = np.arange(2)
width = 0.4

# De eerste grafiek (links) in het figuur
ax[0].bar(x + width / 2, malignant_counts.values, width=width, alpha=0.5, color="red", label="Malignant")
ax[0].bar(x - width / 2, benign_counts.values, width=width, alpha=0.5, color="blue", label="Benign")
ax[0].set_xticks(x)
ax[0].set_xticklabels(["No", "Yes"])
ax[0].set_ylabel("Occurrences")
ax[0].set_xlabel("Metastasis")

# De tweede grafiek (rechts) in het figuur
custom_colors = {"Benign": "blue", "Malignant": "red"}
meta_prop = breast_cancer_df.groupby("Diagnosis Result")["Metastasis"].mean().reset_index()
sns.barplot(data=meta_prop, x="Diagnosis Result", y="Metastasis", hue="Diagnosis Result", alpha=0.5, palette=custom_colors, legend=False)

# Het aanmaken en plaatsen van de legende
handles, labels = ax[0].get_legend_handles_labels()
fig.legend(handles, labels, loc="lower center", ncol=2)

# Het weergeven van het figuur
plt.tight_layout()
plt.show()

# X2-test
# H0: Er is geen verschil in metastasis en tussen goed- en kwaardaardige tumoren
# H1: Er is een verschil in metastasis en tussen goed- en kwaardaardige tumoren
metastasis_table = pd.crosstab(breast_cancer_df["Metastasis"], breast_cancer_df["Diagnosis Result"])
chi2, p_value, dof, expected = chi2_contingency(metastasis_table)
print("p-value:", p_value)
print("Chi²:", chi2)
print("Degrees of freedom:", dof)

# Effect size berekenen
n = metastasis_table.values.sum()
k = min(metastasis_table.shape)
cramers_v = np.sqrt(chi2 / (n * (k - 1)))

print("Cramerss V:", cramers_v)
```

De chi-kwadraattoets toont een significant verband tussen metastase en diagnose (X² = 110.508, p =   <0.0001). Cramer's V (V = 0.736) laat zien dat dit verband groot van sterkte is. Metastase komt vaker voor bij kwaadaardige tumoren.

## Positie
Er is gekeken naar de positie van de tumoren in de borst, hier is de dataset opgedeeld in 4 kwadranten, boven-binnekant, boven-buitenkant, onder-binnekant, onder-buitekant.

```python
# Het aanmaken van nieuwe DataFrames er kwadrant
breast_up_in = breast_cancer_df.query("`Breast Quadrant` == 'Upper inner'")
breast_up_out = breast_cancer_df.query("`Breast Quadrant` == 'Upper outer'")
breast_l_out = breast_cancer_df.query("`Breast Quadrant` == 'Lower outer'")
breast_l_in = breast_cancer_df.query("`Breast Quadrant` == 'Lower inner'")

# Argumenten voor de pie-charts
labels = breast_cancer_df['Diagnosis Result'].unique()
color = ["red", "blue"]
wp = {'alpha':0.5}
pct = autopct="%1.1f%%"

# Het aanmaken van het figuur met titel
fig, ax = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle("Breast Quadrant by Diagnosis", fontsize=14)

# De 4 pie-charts aanmmaken.
ax[0, 0].pie(breast_up_in["Diagnosis Result"].value_counts(), startangle=90, colors=color, wedgeprops=wp, labels=labels, autopct=pct)
ax[0, 0].set_title("Upper Inner Breast Quadrant")
ax[0, 1].pie(breast_up_out["Diagnosis Result"].value_counts(), startangle=90, colors=color, wedgeprops=wp, labels=labels, autopct=pct)
ax[0, 1].set_title("Upper Outer Breast Quadrant")
ax[1, 0].pie(breast_l_in["Diagnosis Result"].value_counts(), startangle=90, colors=color, wedgeprops=wp, labels=labels, autopct=pct)
ax[1, 0].set_title("Lower Inner Breast Quadrant")
ax[1, 1].pie(breast_l_out["Diagnosis Result"].value_counts(), startangle=90, colors=color, wedgeprops=wp, labels=labels, autopct=pct)
ax[1, 1].set_title("Lower Outer Breast Quadrant")

# Het figuur weergeven
plt.show()
```

## Geschiedenis
Er is gekeken naar relatie tussen de geschiedenis van patient (heeft de patient/familielid eerder borstkanker gehad) en of de tumor goedaardige of kwaadaardig is.

```python
# Het groeperen en tellen van de geschiedenis
no_history = breast_cancer_df.query("History == 0").groupby("Diagnosis Result")["History"].count()
yes_history = breast_cancer_df.query("History == 1").groupby("Diagnosis Result")["History"].count()

# Het aanmaken van het figuur met titel
fig, ax = plt.subplots(1, 2, figsize=(15,5))
fig.suptitle("History by Diagnosis", fontsize=14)

# De 2 pie-charts
ax[0].pie(no_history, startangle=90, colors=color, wedgeprops=wp, autopct=pct, labels=labels)
ax[0].set_title("History Absent")
ax[1].pie(yes_history, startangle=90, colors=color, wedgeprops=wp, autopct=pct, labels=labels)
ax[1].set_title("History Present")

# Het figuur weergeven
plt.show()

# X2-test
# H0: Er is geen verschil in geschiedenis en tussen goed- en kwaardaardige tumoren
# H1: Er is een verschil in geschiedenis en tussen goed- en kwaardaardige tumoren
history_table = pd.crosstab(breast_cancer_df["History"], breast_cancer_df["Diagnosis Result"])
chi2, p_value, dof, expected = chi2_contingency(history_table)
print("p-value:", p_value)
print("Chi²:", chi2)
print("Degrees of freedom:", dof)

n = history_table.values.sum()
k = min(history_table.shape)
cramers_v = np.sqrt(chi2 / (n * (k - 1)))

print("Cramers V:", cramers_v)
```

Er is een significant verband gevonden tussen voorgeschiedenis van borstkanker en diagnose (X² = 7.79, p = 0.0053). De effectgrootte (Cramer’s V = 0.195) wijst op een klein maar aanwezig verband: patiënten met een positieve voorgeschiedenis hebben relatief vaker een kwaadaardige tumor dan patiënten zonder voorgeschiedenis.

## Metastasis en Tumor Grootte

Als laatste is de tumor grootte vergeleken tussen patienten met metastasen en patienten zonder metastasen.

```python
no_metastasis = breast_cancer_df.query("Metastasis == 0")
yes_metastasis = breast_cancer_df.query("Metastasis == 1")

fig, ax = plt.subplots(1, 3, figsize=(15,5))
fig.suptitle("Metastasis by Tumor Size", fontsize=14)

ax[0].hist(no_metastasis["Tumor Size (cm)"], bins=7, alpha=0.5, color="blue", label = "No metastasis")
ax[0].set_ylabel("Occurences")
ax[0].set_xlabel("Tumor Size (cm)")
ax[1].hist(yes_metastasis["Tumor Size (cm)"], bins=7, alpha=0.5, color="red", label = "Metastasis")
ax[1].set_ylabel("Metastasis occurences")
ax[1].set_xlabel("Tumor Size (cm)")

sns.kdeplot(no_metastasis["Tumor Size (cm)"], color="blue", ax=ax[2], fill=True, linewidth=2)
sns.kdeplot(yes_metastasis["Tumor Size (cm)"], color="red", ax=ax[2], fill=True, linewidth=2)

# U-test
# H0: Er is geen verschil in tumor grootte en metastasen
# H1: Er is een verschil in tumor grootte en metastasen
no_metastasis_size = no_metastasis["Tumor Size (cm)"].astype(int)
yes_metastasis_size = yes_metastasis["Tumor Size (cm)"].astype(int)

u_stat, p_value = mannwhitneyu(no_metastasis_size, yes_metastasis_size, alternative="two-sided")
print("p-value:", p_value)
print("u-stat:", u_stat)

# Effect size berekenen
n1, n2 = len(no_metastasis_size), len(yes_metastasis_size)
rank_biserial = 1 - (2 * u_stat) / (n1 * n2)
print("Rank-biserial correlation:", rank_biserial)
```

De tumorgrootte verschilt sterk tussen patiënten met en zonder metastasen (U = 789, p = <0.0001). De effectgrootte is groot (r_rb = 0.832) en geeft aan dat metastatische tumoren gemiddeld groter zijn dan niet-metastatische tumoren.
